```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import os
import pandas as pd
import numpy as np
import ast
```

```python
data = pd.read_csv('/Users/arina/study/ds/project/food_recommendation/bbc_goodfood/project_food/data/csv_dataframe/data_goodfood_bbc_parsed.csv', sep='\t')
```

```python
ingredients_data = data['ingredients_parsed'].tolist()
names = data['name'].tolist()

# for id, ingredient in enumerate(ingredients):
    # print([word for single_ingredient in ast.literal_eval(ingredient) for word in single_ingredient.split()])

# Process the input for the doc2vec model
tagged_data = [
                    TaggedDocument( 
                                    words=[
                                            word 
                                            for ingredient in ast.literal_eval(ingredients) 
                                            for word in ingredient.split()
                                        ], 
                                    tags=[str(index)]
                                ) 
                    for index, ingredients in enumerate(ingredients_data)]

# # Define hyperparameters
max_epochs = 20
vec_size = 50
alpha = 0.025

# # Model setup
model_embedding = Doc2Vec(vector_size=100, window=5, min_count=1, dm=1)
model_embedding.build_vocab(tagged_data)
model_embedding.train(tagged_data, total_examples=model_embedding.corpus_count, epochs=20)


# for epoch in range(max_epochs):
#         print('iteration {0}'.format(epoch))
#         model_embedding.train(tagged_data,
#                     total_examples=model_embedding.corpus_count,
#                     epochs=model_embedding.epochs)
        # # decrease the learning rate
        # model_embedding.alpha -= 0.0002
        # # fix the learning rate, no decay
        # model_embedding.min_alpha = model_embedding.alpha
    
```

```python
embeddings = model_embedding.infer_vector(['milk', 'cinnamon', 'oil', 'egg', 'sugar'])
best_recipes = model_embedding.dv.most_similar([embeddings], topn=5)

print(best_recipes)
for value in best_recipes:
    # print(data.iloc[int(value[0])])
    print(int(value[0]))
```
